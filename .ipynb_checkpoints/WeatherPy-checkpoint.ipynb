```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from config import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "Output/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
#consistent results of 601
#np.random.seed(30)
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    #also add country codes to aid analysis
    country_code = citipy.nearest_city(lat_lng[0], lat_lng[1]).country_code
    
    # If the city is unique, then add it to a our cities list
    if (city,country_code) not in cities:
        cities.append((city,country_code))

# Print the city count to confirm sufficient count
len(cities)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=lats,y=lngs,c=(0.7,0.7,0.7))
plt.ylim(-190,190)
plt.xlim(-95,95)
plt.title("Random generated coordinates")
plt.show()
```

## Perform API Calls

```python
# OpenWeatherMap API Key
api_key = weather_api_key

weather_data = []
counter = 1


# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 

###

print("Retrieving Data")
print("------------------------------")

for city in cities:
    # Build query URL
    query_url = url+"&q="+city[0]+","+city[1]
    
    print(query_url)
    
    # Request json file and decrypt it, using a Try / Except Block to avoid errors
    
    try:
        request = requests.get(query_url).json()
        
        #Store information from the city's json into the weather data list
        weather_data.append({"City":request['name'],\
                             "Country":request['sys']['country'],\
                             "Maximum Temperature (F)":request['main']['temp_max'],\
                             "Humidity (%)":request['main']['humidity'],\
                             "Cloud Coverage (%)":request['clouds']['all'],\
                             "Latitude":request['coord']['lat'],\
                             "Longitude":request['coord']['lon'],\
                             "Wind Speed (mph)":request['wind']['speed'],})
    
    # Print success message, with the number of the current city and its name
        print(f"Successfully stored information for City #{counter}: {city}")
    
    # Finish exception with an error message
    except Exception:
        print(f"Error for City #{counter}: {city}")
        print(query_url)
        
    counter+=1
    
    if counter == 101:
        time.sleep(1)
        
print("------------------------------")
print("Data Retrieval Complete")
print("------------------------------")
```

```python
# Convert the list of dictionaries Weather Data to a pandas DataFrame
weather_df = pd.DataFrame(weather_data)

#preview
weather_df.head()
```

```python
#count
weather_df.count()
```

```python
# export result to csv file
weather_df.to_csv(output_data_file)
```

```python
#PLOTS

current_time=time.strftime('%Y/%m/%d %H:%M',time.localtime())

plt.style.use("ggplot")

for i in ("Maximum Temperature (F)","Humidity (%)","Cloud Coverage (%)","Wind Speed (mph)"):
    # make plot
    weather_df.plot(kind="scatter",x="Latitude",y=i,figsize=(10,6),style="o", s=40,edgecolors="b",alpha=0.5)
    plt.title(f"{i} vs. Latitude ({current_time})")
    plt.xlim(-95,95)
    
    # save plot as png file
    plt.savefig(f"output/{i} vs. Latitude.png")
```
